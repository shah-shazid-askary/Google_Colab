```python
def add(x, y):
  """Adds two numbers."""
  return x + y

def subtract(x, y):
  """Subtracts the second number from the first."""
  return x - y

def multiply(x, y):
  """Multiplies two numbers."""
  return x * y

def divide(x, y):
  """Divides the first number by the second."""
  if y == 0:
    return "Error: Division by zero"
  return x / y



```

```python
try:
    num_input1 = float(input("Enter the first number: "))
    num_input2 = float(input("Enter the second number: "))

    print(f"{num_input1} + {num_input2} = {add(num_input1, num_input2)}")
    print(f"{num_input1} - {num_input2} = {subtract(num_input1, num_input2)}")
    print(f"{num_input1} * {num_input2} = {multiply(num_input1, num_input2)}")
    print(f"{num_input1} / {num_input2} = {divide(num_input1, num_input2)}")
except ValueError:
    print("Invalid input. Please enter valid numbers.")
```
