Downword hill climb



```python
def f(x):
  return -x**2+ 4*x;
```

```python
def hill_climbing(f, x0, step_size=0.1, max_iterations=100):
    x = x0  # Start with the initial solution
    steps = [(0, x, f(x))] # Store steps for plotting (iteration, x, f(x))
    for i in range(max_iterations):
        print(f"Iteration {i+1}: x = {x:.2f}, f(x) = {f(x):.2f}")

        # Generate neighbors
        neighbor_left = x - step_size
        neighbor_right = x + step_size

        # Evaluate neighbors
        f_current = f(x)
        f_left = f(neighbor_left)
        f_right = f(neighbor_right)

        # Find the best neighbor
        if f_left < f_current:
            x = neighbor_left  # Move to the left neighbor
        elif f_right < f_current:
            x = neighbor_right  # Move to the right neighbor
        else:
            break  # No better neighbor found, terminate

        steps.append((i+1, x, f(x))) # Store step

    return x, f(x), steps # Return best solution, score, and steps

# Shazid ID:0242310005431201
# Run Hill Climbing
initial_solution = 4.0  # Start at x = 3.0
best_solution, best_score, steps = hill_climbing(f, initial_solution)
print(f"\nBest solution: x = {best_solution:.2f}, f(x) = {best_score:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for plotting the objective function
x_values = np.linspace(-10, 10, 400)
y_values = f(x_values) # Changed to use f(x)

# Plot the objective function
plt.figure(figsize=(10, 6)) # Changed figure size back
plt.plot(x_values, y_values, label='Objective Function')
# Shazid ID:0242310005431201
# Extract x and y values from the steps
step_x = [step[1] for step in steps]
step_y = [step[2] for step in steps]

# Plot the steps taken by the algorithm
plt.scatter(step_x, step_y, color='red', label='Hill Climbing Steps')
plt.plot(step_x, step_y, color='red', linestyle='--', alpha=0.5) # Connect the steps

# Mark the starting and ending points
plt.scatter(steps[0][1], steps[0][2], color='green', marker='o', s=100, label='Start')
plt.scatter(best_solution, best_score, color='purple', marker='*', s=200, label='Best Solution Found') # Changed marker size back

plt.ylim(0, max(y_values) * 1.1) # Set y-axis to start from 0
plt.title('Hill Climbing Optimization Process')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
