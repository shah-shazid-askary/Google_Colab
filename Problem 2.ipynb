```python
class LibraryItem:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
        self._is_checked_out = False # Internal state

    def check_out(self):
        if not self._is_checked_out:
            self._is_checked_out = True
            print(f"'{self.title}' by {self.author} has been checked out.")
        else:
            print(f"'{self.title}' is already checked out.")

    def return_item(self):
        if self._is_checked_out:
            self._is_checked_out = False
            print(f"'{self.title}' by {self.author} has been returned.")
        else:
            print(f"'{self.title}' was not checked out.")

    def __str__(self):
        status = "checked out" if self._is_checked_out else "available"
        return f"Title: {self.title}, Author: {self.author}, Year: {self.year}, Status: {status}"

class Magazine(LibraryItem):
    def __init__(self, title, author, year, issue_number):
        super().__init__(title, author, year)
        self.issue_number = issue_number

    def display_issue(self):
        print(f"This is issue number {self.issue_number} of '{self.title}'.")

    def __str__(self):
        status = "checked out" if self._is_checked_out else "available"
        return f"Title: {self.title}, Author: {self.author}, Year: {self.year}, Issue: {self.issue_number}, Status: {status}"
```

```python
# Create objects of both classes
book = LibraryItem("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 1979)
magazine = Magazine("National Geographic", "National Geographic Society", 2023, 10)

# Demonstrate the use of methods
print("--- Book Actions ---")
print(book)
book.check_out()
print(book)
book.check_out() # Try checking out again
book.return_item()
print(book)
book.return_item() # Try returning again

print("\n--- Magazine Actions ---")
print(magazine)
magazine.check_out()
print(magazine)
magazine.return_item()
print(magazine)
magazine.display_issue()
```
