```python
def f(x):
    return x**2 + 4*x + 4
def df(x):
    return 2*x + 4
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    for i in range(iterations):
        x = x - learning_rate * df(x)  # update step
        print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
    return x

starting_point = 0
learning_rate = 0.1
iterations = 10

minimum = gradient_descent(starting_point, learning_rate, iterations)
print(f"\nLocal minimum occurs at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Redefine the gradient_descent function to store history for plotting
def gradient_descent_with_history(starting_point, learning_rate, iterations):
    x_history = [starting_point]
    f_x_history = [f(starting_point)]
    x = starting_point
    for _ in range(iterations):
        x = x - learning_rate * df(x)
        x_history.append(x)
        f_x_history.append(f(x))
    return x_history, f_x_history

# Get the history of x and f(x) values
x_history, f_x_history = gradient_descent_with_history(starting_point, learning_rate, iterations)

# Generate x values for plotting the function curve
x_vals = np.linspace(min(min(x_history), -5), max(max(x_history), 1), 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2 + 4x + 4', color='blue')
plt.scatter(x_history, f_x_history, color='red', s=50, zorder=5, label='Gradient Descent Path')
plt.plot(x_history, f_x_history, color='red', linestyle='--', linewidth=1, alpha=0.7)

# Mark the starting point and the final minimum
plt.scatter(x_history[0], f_x_history[0], color='green', marker='o', s=100, label='Starting Point')
plt.scatter(x_history[-1], f_x_history[-1], color='purple', marker='x', s=100, label='Final Minimum')

plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
